### Matrix multiplication

Two may ways of performing multiplication in neural networs and deep learning:

1. Element-wise multiplicaation
2. Matrix Multiplicaion (dot product)


1. The **inner dimensions** must math:
* ` (3, 2) @ (3, 2)` won't work
* ` (2, 3) @ (3, 2)` will work
* ` (3, 2) @ (2, 3)` will work
2. The resulting matrix has the shape of the **outer dimensions**:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

```python
# Element wise multiplication
print(tensor, "*", tensor)
print(f"Equals: {tensor * tensor}")
```

```python
# Matrix multiplication
torch.matmul(tensor, tensor)
```

```python
%%time
value = 0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]
print(value)
```

## Reshaping, stacking, squeezing and unsqueezing tensors
* Reshaping - reshapes an input tonsor to a defined shape
* View - Return a view of an input tensor of certain shape but keep the memory as the original tensor
* Stacking - Combine multiple tensors on top of each other (vstack) or side by side (hstack)
* Squeeze - remove all `1` dimension from a tensor
* Unsqueeze - add a `1` dimension to a target tensor
* Permute - Return a view of the input with dimensions permuted (swapped) in a certain way

```python
# create a tensor
import torch
x = torch.arange(1., 10.)
x, x.shape
```

```python
# Add ann extra dimension
x_reshaped = x.reshape(1, 9)
x_reshaped, x_reshaped.shape
```

```python
z = x.view(1, 9)
z, z.shape
```

```python
# Changing z changes(because a view of a tensor shares the same memory as the original input)
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x, x, x], dim=1)
x_stacked
```

```python
# torch.squeeze
x_reshaped.squeeze()
```

```python
# torch.permute - rearranges the dimensions of a target tensor in a specified order
x_original = torch.rand(size=(224, 224, 3))# [height, width, colour_channels]

# Permute the oriingal tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) #shift 0->1, 1->2, 2->0

print(f"Previous shape: {x_original.shape}")
print(f"New shape:{x_permuted.shape}")
```
